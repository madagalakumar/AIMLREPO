```python
from transformers import pipeline
```

```python
qa_model = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad", tokenizer="distilbert-base-uncased")
```

```python
context = "Artificial intelligence, commonly known as AI and often abbreviated as such, pertains to the advancement of computer systems with the ability to execute tasks typically demanding human intelligence."
question = "what is artificial intelligence and its goals?"

result = qa_model(question=question, context=context)
print(f"Answer: '{result['answer']}'")
print(f"Confidence: {result['score']}")
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-de")
```

```python
text_to_translate = "Hlo jagadeesh, plz save me from this situation."

translated_text = translator(text_to_translate, max_length=40)[0]['translation_text']
print(f"Translated text: {translated_text}")
```
