```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv("cast.csv")
cast_df=pd.DataFrame(d)
```

```python
cast_df
```

```python
d = pd.read_csv("release_dates.csv")

release_dates_df = pd.DataFrame(d)
```

```python
release_dates_df
```

# 1.Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
christmas_movies_usa = release_dates_df[(release_dates_df['title'].str.contains("Christmas", case=False, na=False)) &
                                         (release_dates_df['country'] == "USA")]

christmas_movies_usa['month'] = pd.to_datetime(christmas_movies_usa['date']).dt.month

monthly_counts = christmas_movies_usa['month'].value_counts().sort_index()

month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Monthly Distribution of Movies with "Christmas" in Title (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.xticks(range(12), month_names, rotation=45)
plt.show()
```

# Make a bar plot of the day of the week on which movies with "Romance" in their title terid to be released in the USA.

# In which months do films with Judi Dench tend to be released in the USA?

```python
judi_dench_movies = cast_df[cast_df['name'] == 'Judi Dench']

judi_dench_movies = pd.merge(judi_dench_movies, release_dates_df, on=['title', 'year'])

judi_dench_movies['date'] = pd.to_datetime(judi_dench_movies['date'], errors='coerce')
judi_dench_movies = judi_dench_movies.dropna(subset=['date'])

judi_dench_movies_usa = judi_dench_movies[judi_dench_movies['country'] == 'USA']

judi_dench_movies_usa = judi_dench_movies_usa.copy()
judi_dench_movies_usa['release_month'] = judi_dench_movies_usa['date'].dt.month

release_month_counts = judi_dench_movies_usa['release_month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
release_month_counts.plot(kind='bar', color='skyblue')
plt.title("Judi Dench Movies Released in the USA by Month")
plt.xlabel("Month")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```

# In which months do films with Tom Cruise tend to be released in the USA?

```python
tom_cruise_movies = cast_df[cast_df['name'] == 'Tom Cruise']

tom_cruise_movies = pd.merge(tom_cruise_movies, release_dates_df, on=['title', 'year'])

tom_cruise_movies_usa = tom_cruise_movies[tom_cruise_movies['country'] == 'USA']

tom_cruise_movies_usa['date'] = pd.to_datetime(tom_cruise_movies_usa['date'], errors='coerce')

tom_cruise_movies_usa['release_month'] = tom_cruise_movies_usa['date'].dt.month

release_month_counts = tom_cruise_movies_usa['release_month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
release_month_counts.plot(kind='bar', color='skyblue')
plt.title("Tom Cruise Movies Released in the USA by Month")
plt.xlabel("Month")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```
