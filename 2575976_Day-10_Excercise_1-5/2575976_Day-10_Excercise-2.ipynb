```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv("titles.csv")
titles_df=pd.DataFrame(d)
```

# 1.what are the ten most common movie names of all time?

```python
title_counts = titles_df['title'].value_counts()

top_ten_titles = title_counts.head(10)

print("Top Ten Most Common Movie Names of All Time:")
print(top_ten_titles)
```

# 2.which three years of the 1930s saw the most films released?

```python
movies_1930s = titles_df[(titles_df['year'] >= 1930) & (titles_df['year'] <= 1939)]

movie_counts_by_year = movies_1930s['year'].value_counts()

top_three_years = movie_counts_by_year.head(3)

print("Top Three Years in the 1930s with the Most Films Released:")
print(top_three_years)
```

# 3.Plot the number of films that have been released each decade over the history of cinema.

```python
titles_df['decade'] = (titles_df['year'] // 10) * 10

films_per_decade = titles_df['decade'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
films_per_decade.plot(kind='bar', color='skyblue')
plt.title('Number of Films Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.xticks(rotation=45)
plt.show()
```

# 15.Plot the n-values of the roles that Judi Dench has played over her career.

```python
judi_dench_roles = cast_df[cast_df['name'] == "Judi Dench"]

judi_dench_roles = judi_dench_roles[pd.to_numeric(judi_dench_roles['n'], errors='coerce').notnull()]

judi_dench_roles['n'] = pd.to_numeric(judi_dench_roles['n'])

plt.figure(figsize=(12, 6))
plt.scatter(judi_dench_roles['year'], judi_dench_roles['n'], alpha=0.5, color='purple')
plt.title('n-Values of Roles Played by Judi Dench Over Her Career')
plt.xlabel('Year')
plt.ylabel('n-Value')
plt.grid(True)
plt.show()
```

# 18.How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
roles_1950s = cast_df[(cast_df['year'] >= 1950) & (cast_df['year'] <= 1959)]

leading_roles = roles_1950s[roles_1950s['n'] == 1]

leading_roles_actors = leading_roles[leading_roles['type'] == 'actor']
leading_roles_actresses = leading_roles[leading_roles['type'] == 'actress']

num_leading_roles_actors = len(leading_roles_actors)
num_leading_roles_actresses = len(leading_roles_actresses)

print(f"Number of leading roles (n=1) for actors in the 1950s: {num_leading_roles_actors}")
print(f"Number of leading roles (n=1) for actresses in the 1950s: {num_leading_roles_actresses}")
```
